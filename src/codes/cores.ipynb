```julia
using Colors
```

```julia
function gen_color()
    letter = num2hex(rand(0:255))[15:16]
    for i in 1:2
        letter = string(letter,num2hex(rand(0:255))[15:16])
    end
    cor = string("#",letter)
    return cor
end
```

```julia
using PyPlot
x = linspace(-1,1,101)
y = sin(x*pi)

plot(x,y,color = cor)
```
