```julia
using PyPlot
```

```julia
phi1(x,e) = (x - e[1])/(e[2] - e[1])  
phi2(x,e) = (e[2] - x )/(e[2] - e[1]) 
```

```julia
e1 = [0,1/3]
e2 = [1/3,2/3]
e3 = [2/3,1]
```

```julia
phi1(x,e1)
```

```julia
subplot(2,1,1)
x = collect(e1[1]:0.01:e1[2])
y = phi1(x,e1)
plot(x,y,color = "red")


x = collect(e1[1]:0.01:e1[2])
y = phi2(x,e1)
plot(x,y,color =  "red")




x = collect(e2[1]:0.01:e2[2])
y = phi1(x,e2)
plot(x,y,color = "green")


x = collect(e2[1]:0.01:e2[2])
y = phi2(x,e2)
plot(x,y,color = "green")



x = collect(e3[1]:0.01:e3[2])
y = phi1(x,e3)
plot(x,y,color = "yellow")


x = collect(e3[1]:0.01:e3[2])
y = phi2(x,e3)
plot(x,y,color = "yellow")

#xlim([0,1])
[axvline(x,color = "black") for x in linspace(0,1,4)]
```
