```julia
using Jacobi
using PyPlot
```

```julia

#intervalo 
a = -5 ; b = 5


#f(x) = 1./(1 + x.^2)
f(x) = sin(x*pi)
Q=  19
nx = 1000
x2 = linspace(a, b, nx)
ye2 = f(x2)
plot(x2, ye2, "-",label="Original")
title("Interpolação \n para raízes equidistantes")
legend()
```

```julia

Q = 50
erroc = 0.0
for Q in 1:1:50
    z = chebyshev_zeros(Q)
    #z = zglj(Q)
    nx= 1000
    x = linspace(a, b, nx)

    #altera o intervalo de interpolação
    z = ((1-z)/(2)).*a + ((1+z)/(2)).*b
    #estabelece as funções
    ye = f(z)
    #interpola os pontos por polinômios de lagrange
    Im = interp_mat(x, z)
    y =  Im * ye
    #aprox
    #plot(x,y)
    erro = maxabs(y-ye2)
    erroc = [erroc maxabs(y-ye2)]
    println("erro para $Q = $erro") 
end
```
